# Preprocesamiento

En la Parte 1 de este taller, abordaremos el primer paso del análisis de texto. Nuestro objetivo es convertir los datos de texto crudos y desordenados en un formato consistente. Este proceso a menudo se denomina **preprocesamiento**, **limpieza de texto **, o **normalización de texto**.

Notarás que, al final del preprocesamiento, nuestros datos siguen en un formato que podemos leer y entender. En las Partes 2 y 3, comenzaremos a convertir los datos de texto en una representación numérica, un formato que puede ser manejado más fácilmente por las computadoras.

## Procesos Comunes

El preprocesamiento no es algo que podamos lograr con una sola línea de código. A menudo comenzamos familiarizándonos con los datos, y en el camino, obtenemos una comprensión más clara de la granularidad del preprocesamiento que queremos aplicar.

Típicamente, comenzamos aplicando un conjunto de procesos comúnmente utilizados para limpiar los datos. Estas operaciones no alteran sustancialmente la forma o el significado de los datos; sirven como un procedimiento estandarizado para reformatear los datos en un formato consistente.

Los siguientes procesos, por ejemplo, se aplican comúnmente para preprocesar textos en inglés de varios géneros. Estas operaciones se pueden realizar utilizando funciones integradas de Python, como métodos de `cadenas`, y Expresiones Regulares.
- Convertir el texto a minúsculas
- Eliminar signos de puntuación
- Eliminar espacios en blanco adicionales
- Eliminar palabras vacías (stop words)

Después del procesamiento inicial, podemos optar por realizar procesos específicos para la tarea, cuyos detalles a menudo dependen de la tarea posterior que queremos realizar y de la naturaleza de los datos de texto (es decir, sus características estilísticas y lingüísticas).
