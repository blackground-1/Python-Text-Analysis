### Importar los Datos de Texto

Los datos de texto con los que trabajaremos están en un archivo CSV. Contiene tweets sobre aerolíneas de EE. UU., extraídos de febrero de 2015.

Leamos el archivo `airline_tweets.csv` en un dataframe con `pandas`.


```python
# Import pandas
import pandas as pd

# File path to data
csv_path = '../data/airline_tweets.csv'

# Specify the separator
tweets = pd.read_csv(csv_path, sep=',')
```

```python
# Show the first five rows
tweets.head()
```

El dataframe tiene una fila por cada tweet. El texto del tweet se encuentra en la columna `text`.  
- `text` (`str`): el texto del tweet.

Otros metadatos que nos interesan incluyen:  
- `airline_sentiment` (`str`): el sentimiento del tweet, etiquetado como "neutral," "positive" o "negative."  
- `airline` (`str`): la aerolínea sobre la que se tuiteó.  
- `retweet count` (`int`): cuántas veces se retuiteó el tweet.  


Echemos un vistazo a algunos de los tweets:

```python
print(tweets['text'].iloc[0])
print(tweets['text'].iloc[1])
print(tweets['text'].iloc[2])
print(tweets['text'].iloc[3])
```

### Conversión a Minúsculas

Aunque reconocemos que el uso de mayúsculas y minúsculas en una palabra aporta información, a menudo trabajamos en contextos donde no podemos aprovechar adecuadamente esa información.

Con mayor frecuencia, el análisis posterior que realizamos es **insensible a mayúsculas y minúsculas**. Por ejemplo, en un análisis de frecuencia, queremos considerar las diferentes formas de la misma palabra. Convertir los datos de texto a minúsculas facilita este proceso y simplifica nuestro análisis.

Podemos lograr fácilmente la conversión a minúsculas con el método de cadenas [`.lower()`](https://docs.python.org/3/library/stdtypes.html#str.lower); consulta la [documentación](https://docs.python.org/3/library/stdtypes.html#string-methods) para más funciones útiles.

Apliquémoslo en el siguiente ejemplo:


```python
# Print the first example
first_example = tweets['text'][108]
print(first_example)
```

```python
# Check if all characters are in lowercase
print(first_example.islower())
print(f"{'=' * 50}")

# Convert it to lowercase
print(first_example.lower())
print(f"{'=' * 50}")

# Convert it to uppercase
print(first_example.upper())
```
